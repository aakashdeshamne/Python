```python
#Getting started with Matplotlib and Seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Basics
x = np.linspace(0, 5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
```

```python
z=x**3
plt.plot(x,z)
```

```python
plt.title("Graph between x and y")
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
#subplots
plt.subplot(2,2,1)
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,z)
plt.subplot(2,2,3)
plt.plot(y,z)
plt.subplot(2,2,4)
plt.plot(z,y)
```

```python
#object oriented method to get full control of axis
x=np.linspace(0,5,11)
y=x**2
z=x**3
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2=fig.add_axes([0.2,0.56,0.3,0.3])
axis2.plot(x,z)
```

```python
#Types of plots
#Scattered plot
plt.scatter(x,y)
```

```python
#Histogram
data=np.random.randn(1000)
plt.hist(data)
```

```python
plt.hist(x)
```

```python
plt.hist(z,bins=20)
```

```python
#box pot
data=np.random.randn(100)
plt.boxplot(data)
```
