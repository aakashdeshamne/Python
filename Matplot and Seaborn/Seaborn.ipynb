```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df=sns.load_dataset('tips')
df.head()
```

```python
# dispot
sns.displot(df['total_bill'],bins=100)
```

```python
sns.displot(df['total_bill'],bins=30,kde=True)
```

```python
# joints plot
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
# lets learn about pair plot
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

```python
sns.pairplot(df,hue="size")
```

```python
# count plot
sns.countplot(df['size'])
```

```python
sns.countplot(x=df['sex'])
```
